```python
ax=sns.countplot(x="Size",data=df,palette="viridis",hue="Size")             #count=count of Size
```

```python
ax=sns.countplot(x="Size",data=df,palette="spring",hue="Size")
for bar in ax.containers:                     #Here,bar is a variable,ax.container
    ax.bar_label(bar)                         #bar_label is to label variables in bar chart
```

Group By:

```python
df.groupby(["Size"],as_index=False)["Quantity"].sum().sort_values(by="Quantity",ascending=False)
```

```python
S_Quantity=df.groupby(["Size"],as_index=False)["Quantity"].sum().sort_values(by="Quantity",ascending=False)
sns.barplot(x="Size",y="Quantity",data=S_Quantity,palette="spring",hue="Size")
```

```python
# Histogram
df["Size"].hist()
```

```python
df["Category"]=df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=10,edgecolor="Black",color="SkyBlue")
plt.xticks(rotation=90)      # .xticks is used to rotate the labels in x axis according to the value given in the rotation
plt.show()
```

Note: From Above Graph you can see that most of the buyers are for T-Shirt

```python
# Checking B2B data by using Pie Chart
B2B_Check=df["B2B"].value_counts()

#Plot the Pie Chart
plt.pie(B2B_Check,labels=B2B_Check.index,autopct="%1.1f%%")
#plt.axis("equal")
plt.show()
```

Note:From above chart we can see that maximum i.e. 99.2% of buyers are retailers and 0.8% are B2B buyers

```python
#Prepare data for Scatter Plot
x_data=df["Category"]
y_data=df["Size"]

#Plot the Scatter Plot
plt.scatter(x_data,y_data)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("Scatter Plot")
plt.show()
```

```python
# Plot count of Cities by State
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="ship-state")
plt.xlabel("Shipping-state")
plt.ylabel("Distribution of State")
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 States
top_10_state=df["ship-state"].value_counts().head(10)
#Plot count of cities by State
plt.figure(figsize=(12,6))
sns.countplot(data=df[df["ship-state"].isin(top_10_state.index)],x="ship-state")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("Distribution of State")
plt.xticks(rotation=45)
plt.show()
```

Note: From above Graph you can see that most of the buyers are from Maharashtra
